```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation

import numpy as np

%matplotlib ipympl

#sine of sine of sine of sine n times
def foo(n,x):
    for i in range(0,n):
        x = np.sin(x)
    return x

x=np.arange(-2*np.pi,2*np.pi,0.05)
y=foo(1,x)
fig, ax = plt.subplots()
ax.set_xlim(0,2*np.pi)
ax.set_ylim(-1.1,1.1)

line, = ax.plot([])
ax.set_title("plot")
ax.set_xlabel('angle')
ax.set_ylabel('sine of sine of sine n-times')

def animate(frame_num):
    y=foo(frame_num,x)
    line.set_data((x,y))
    return line

anim = FuncAnimation(fig,animate, frames=10000, interval = 50)
plt.show()


            
```
